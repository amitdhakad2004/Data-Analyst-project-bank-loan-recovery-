```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Hp\\Downloads\\bank_loan_recovery_dataset.csv")
```

```python
df.columns
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

```

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x= 'repayment_status', order=df[ 'repayment_status'].value_counts().index)
ax.bar_label(ax.containers[0])
plt.title('repayment_status')                  
plt.show()
```

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x= 'payment_method', order=df[ 'payment_method'].value_counts().index)
ax.bar_label(ax.containers[0])
plt.title('payment_method')                  
plt.show()
```
